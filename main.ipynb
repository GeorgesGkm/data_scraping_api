```python
import requests
import pandas as pd 
```

```python
# username de Github
username = '#####'

#URL de l'API pour la recuperer  les depots d'un utilisateur
url = f"https://api.github.com/users/{username}/repos"
```

```python
#Effectuer une requete GEt pour récupérer les données de l'api
response = requests.get(url)

if response.status_code == 200:
    repos_data = response.json()
    
    repo_list = []
    for repo in repos_data:
        repo_info = {
            'Name' : repo['name'],
            'Description' : repo['description'],
            'Start' : repo['stargazers_count'],
            'Forks' : repo['forks_count'],
            'Language' : repo['language'],
            'URL' : repo['url']
        }
        repo_list.append(repo_info)
        
    df = pd.DataFrame(repo_list)
    
    print(df.head())
else:
    print(f"la requete à echoué avec le statut : {response.status_code}")
```
